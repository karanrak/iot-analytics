```python
import pandas as pd
import random
import math


def runsim(df,ri,nri,rs,nrs):
    rt_int = ri
    nonrt_int = nri
    rts = rs
    nonrts = nrs
    while True:
        lrow = df.iloc[[-1]]
        nrow = df.iloc[[-1]]
        if int(lrow.mc)>=timelimit:
            break
        if (int(lrow.rtcl) <= int(lrow.nonrtcl)) & (int(lrow.rtcl) <= int(lrow.scl)):
            nrow.mc = lrow.rtcl
            nrow.rtcl += rt_int
            rt_int = ri * -math.log(random.random())
            if int(lrow.s)==0:
                nrow.s=1
                nrow.scl= nrow.mc + rts
                rts = rs * -math.log(random.random())
            elif int(lrow.s)==1 and int(nrow.mc) == int(lrow.scl):
                nrow.scl= nrow.mc + rts
                rts = rs * -math.log(random.random())
            elif int(lrow.s)==1:
                nrow.nrt += 1
            elif int(lrow.s)==2:
                nrow.preempt = lrow.scl-lrow.rtcl
                nrow.s = 1
                nrow.scl= nrow.mc + rts
                rts = rs * -math.log(random.random())
                if int(nrow.preempt > 0):
                    nrow.nnonrt += 1
        elif int(lrow.nonrtcl) <= int(lrow.scl):
            nrow.mc = lrow.nonrtcl
            nrow.nonrtcl += nonrt_int
            nonrt_int = nri * -math.log(random.random())
            if int(lrow.s)==0:
                nrow.s=2
                nrow.scl= nrow.mc + nonrts
                nonrts = nrs * -math.log(random.random())
            elif int(nrow.mc) == int(lrow.scl):
                if int(lrow.nrt) > 0:
                    nrow.nrt -= 1
                    nrow.s = 1
                    nrow.scl= nrow.mc + rts
                    rts = rs * -math.log(random.random())
                elif int(lrow.preempt) > 0:
                    nrow.scl = lrow.scl + lrow.preempt
                    nrow.s=2
                    nrow.preempt = 0
                else:    
                    nrow.scl= nrow.mc + nonrts
                    nonrts = nrs * -math.log(random.random())
            elif int(lrow.s)==1:
                nrow.nnonrt += 1
            elif int(lrow.s)==2:
                nrow.nnonrt += 1
        else:
            nrow.mc = lrow.scl
            if int(lrow.nrt) > 0:
                nrow.scl = nrow.mc + rts
                rts = rs * -math.log(random.random())
                nrow.nrt -= 1
                nrow.s = 1
            elif int(lrow.preempt) > 0:
                nrow.scl = nrow.mc + lrow.preempt
                nrow.nnonrt -= 1
                nrow.s = 2
                nrow.preempt = 0
            elif int(lrow.nnonrt) > 0:
                nrow.scl = nrow.mc + nonrts
                nonrts = nrs * -math.log(random.random())
                nrow.nnonrt -= 1
                nrow.s = 2
            else:
                nrow.s = 0
                nrow.scl = 9999999                
        df = df.append(nrow)
    return df

def write_file(df,fileno):
    writer = pd.ExcelWriter('task_2_'+fileno+'.xlsx', engine='xlsxwriter')

    df2.to_excel(writer, sheet_name='Sheet1')

    writer.save()
```
